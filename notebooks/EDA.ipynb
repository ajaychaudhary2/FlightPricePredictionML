```python

import numpy as np
# Function to cyclically encode the time
def encode_time(time_str):
    hours, minutes = map(int, time_str.split(':'))
    total_minutes = hours * 60 + minutes
    return (np.sin(2 * np.pi * total_minutes / 1440),  # 1440 minutes in a day
            np.cos(2 * np.pi * total_minutes / 1440))

# Apply the function to create sine and cosine columns
df[['Dep_Time_Sin', 'Dep_Time_Cos']] = df['Dep_Time'].apply(encode_time).apply(pd.Series)
```

```python
def convert_duration(duration):
    total_minutes = 0
    parts = duration.split()  # Split by space
    for part in parts:
        if 'h' in part:  # Extract hours
            hours = int(part.split('h')[0])
            total_minutes += hours * 60
        elif 'm' in part:  # Extract minutes
            minutes = int(part.split('m')[0])
            total_minutes += minutes
    return total_minutes

# Apply the function to the Duration column
df['Total_Duration_Minutes'] = df['Duration'].apply(convert_duration)
```

```python
df.head()
```

```python
df=df.drop(labels=["Duration"],axis=1)
```

```python
def time_to_minutes(time_str):
    time_parts = time_str.split(':')
    hours = int(time_parts[0])
    minutes = int(time_parts[1])
    return hours * 60 + minutes

df['Arrival_Time_Minutes'] = df['Arrival_Time'].str.split(' ').str[0].apply(time_to_minutes)
df['Arrival_Time_Sin'] = np.sin(2 * np.pi * df['Arrival_Time_Minutes'] / 1440)  # 1440 minutes in a day
df['Arrival_Time_Cos'] = np.cos(2 * np.pi * df['Arrival_Time_Minutes'] / 1440)
```
